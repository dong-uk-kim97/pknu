### 데이터 통합 : 행단위 및 열단위 통합 
- concat : 열로 합칠 때 행의 갯수가 맞아야 함 / 행은 관계 없음

- axis : 행과 열을 구분하는 속성(0은 행, 1은 열을 의미함)

```python
### 샘플 데이터 생성하기
df_temp1 = pd.DataFrame([['Dog',3],['Cat',5],['Tiger',2]],
                       index=[0,1,2],
                       columns=['동물','나이'])

df_temp2 = pd.DataFrame([['집',0],['집',0],['초원',2]],
                       index=[0,1,2],
                       columns=['사는곳','뿔갯수'])

df_temp1
df_temp2

### 행단위 데이터 통합하기
pd.concat([df_temp1,df_temp2],axis=0,ignore_index=True)
```

```python
### 열단위 데이터 통합하기
pd.concat([df_temp1,df_temp2],axis=1)
```
