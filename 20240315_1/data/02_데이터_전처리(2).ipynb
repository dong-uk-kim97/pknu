```python
### 라이브러리 정의
import pandas as pd
```

```python
### 결측치 데이터 확인하기
# - isnull() : 결측치가 있는 값들을 True로, 아닌 경우 False로 표시됨 
df.isnull()
```

```python
### 결측치가 아닌 데이터
# - notnull() : 결측치가 아닌 값들을 True로, 결측치인 경우 False로 표시됨 
df.notnull()
```

```python
### 컬럼 별 결측치 현황 확인하기
# - sum() : 각 컬럼의 행 전체를 sum
df.isnull().sum()
```


<결측치 처리 방법>
 1. 결측치가 있는 부분의 데이터를 사용할지/말지 결정

 2. 사용 안하면
    
    - 컬럼과 행 중에 어느 부분을 제거할지 결정
      
 3. 사용 한다면
    - 대체하는 방법을 사용
      
      -- 숫자 데이터인 경우 : 해당 컬럼 전체값의 평균으로 대체
      
      -- 범주형 데이터인 경우 : 범주의 비율 대비 대체
      
      -- 무조건 0으로 대체
      
      -- 결측 데이터가 속한 행의 컬럼들과 유사한 다른 행의 데이터들의 평균으로 대체
      
      -- 결측 데이터가 속한 컬럼의 직전/직후 행의 평균으로 대체

```python
### 결측치가 포함된 모든 컬럼 삭제하기
df_drop_allcolumn=df.dropna(axis=1)
df_drop_allcolumn.info()
```

```python
## 특정 컬럼의 결측치 데이터만 삭제하기
df['대여거치대'].dropna(axis=0)
```

```python
df.info()
```

```python
### 여러 개의 컬럼의 결측치가 있는 행 제거하기
# - 대여거치대, 대여거치대, 이용시간 데이터 중에 결측치가 있는 행을 삭제
df[['대여거치대', '대여거치대', '이용시간']].dropna(axis=0)
```

```python
### 이용 거리 데이터 중에 결측치가 있는 부분을 0으로 대체하기
df['이용거리'].fillna(0)
# df['이용거리'] = df['이용거리'].fillna(0)
```

```python
df.info()
```

```python
### 평균으로 대체하기
# - 이용거리의 결측 데이터들을 평균으로 대체하기
df['이용거리'].fillna(df['이용거리'].mean())
```

```python
### 결측치가 있는 모든 컬럼들은 각 컬럼의 평균으로 대체하여 반영시키기
df['이용거리']=df['이용거리'].fillna(df['이용거리'].mean())
df['이용시간']=df['이용시간'].fillna(df['이용시간'].mean())
df['대여거치대']=df['대여거치대'].fillna(df['대여거치대'].mean())
df['반납거치대']=df['반납거치대'].fillna(df['반납거치대'].mean())
df.isnull().sum()
```

```python
### 나이 데이터 중에 100세 이상 데이터와 아닌 데이터의 현황(빈도)를 조회
(df['나이']>=100).value_counts()
```

```python
### 넘파이(numpy) 라이브러리 정의
import numpy as np
```

```python
### numpy 함수를 이용해서 25% 및 75% 시점의 4분위 수 확인
# - 결괏값은 배열(리스트 아님) 형태로 2개의 값을 return함
q1, q3 = np.percentile(df['나이'],[25,75])
q1, q3
```

<이상치 계산공식>    
 - IQR = q3 - q1
 - Max Outlier(최대) = q3 + (IQR * 1.5)
 - Min Outlier(최소) = q1 - (IQR * 1.5)

![](./image.png)

```python
iqr = q3 - q1
iqr
```

```python
### Max Outlier 계산
upper_bound = q3 + (iqr * 1.5)
upper_bound
```

```python
### Min Outlier 계산
lower_bound = q1 - (iqr * 1.5)
lower_bound
```

```python
### upper_bound와 lower_bound의 기준값을 초과하는 데이터 전체를 조회
df[(df['나이']>upper_bound)|(df['나이']<lower_bound)]
```

```python
### upper_bound와 lower_bound의 기준값을 초과하는 데이터 전체를 조회 / 강사님 코드
c = (df['나이']>upper_bound)|(df['나이']<lower_bound)
df[c]
```

```python
### 이상치 10건을 해당 컬럼의 평균값으로 대체하기
# - 평균값 계산하기(이상치를 제외한 평균을 계산하여 타당한 것 같음)
age_mean = int(df[c==False]['나이'].mean())
age_mean
```

```python
### 데이터프레임 복제 방식 2가지
# - 첫번째 방식
#  -- df_temp에서 변형(추가,수정,삭제 등등)의 결과는 df에도 동일하게 적용
#    --> df_temp와 df는 같은 메모리 공간(같은 메모리 주소)을 바라본다.
#    --> df_temp는 df의 메모리 주소를 가진다.    
# df_temp = df

# - 두 번째 방식
#   -- copy()를 이용하게 되면, 새로운 메모리 공간을 만들어서 새로운 메모리 주소가 생성됨
#     --> df_temp는 새로운 메모리 주소를 가진다.
#df_temp = df.copy()

### v = 1 (대입 연산자)
### v = pd.DataFrame() (할당 연산자)
### 점(`.`) : 접근 지정자
```

```python
### 나이 컬럼의 이상치 10건에 대해서 위의 평균값으로 대체하기
df_temp = df.copy()
c = (df_temp['나이']>upper_bound)|(df_temp['나이']<lower_bound)
df_temp.loc[c,'나이']=age_mean # 강사님 코드
df_temp
```
